## Validating Input

```python
# validate SAT user input, divisible by 10 and >=400 and <= 1600
def validate_SAT(user_SAT_str):
    valid = True
    
    #Validate <0> or "prefer not to say" input
    if str(user_SAT_str) == "<0>":
        return True
    
    if not (user_SAT_str.isdigit() and int(user_SAT_str) >= 0 and int(user_SAT_str) <= 1600):
        valid = False
    if int(user_SAT_str) % 10 != 0 or int(user_SAT_str) < 400:
        valid = False
    return valid

# validate ACT user input, <= 36
def validate_ACT(user_ACT_str):
    #Validate <0> or "prefer not to say" input
    if str(user_ACT_str) == "<0>":
        return True
    else:
        return user_ACT_str.isdigit() and int(user_ACT_str) >=0 and int(user_ACT_str) <=36

def validate_GPA(user_GPA_str):
    #Validate <0> or "prefer not to say" input
    if str(user_GPA_str) == "<0>":
        return True
    else:
        return (user_GPA_str.replace('.', '', 1).isdigit() or user_GPA_str.isdigit()) and ((float(user_GPA_str) >=0 and float(user_GPA_str) <= 4.0)) or (int(user_GPA_str) >= 0 and int(user_GPA_str) <= 4)

# make sure each weight is a number between 0-100
def validate_weight(weight):
    if str(weight).isnumeric() == False:
        return False
    if int(weight) > 100 or int(weight) < 0:
        return False
    return True

# make sure weights are numbers and add up to 100
def validate_weights(weight1, weight2, weight3, weight4, weight5):
    weights = [weight1, weight2, weight3, weight4, weight5]
    for weight in weights:
        if str(weight).isnumeric() == False:
            return False
    if int(weight1) + int(weight2) + int(weight3) + int(weight4) + int(weight5) != 100:
        return False
    return True

# make sure input state is a valid state abbreviation
def validate_user_state(user_state):
    if user_state not in states_abb and user_state != 'N/A':
        return False
    return True
```

```python
# Tuition Match
# separates colleges into in-state vs out-of-state for the user based on the state they live in
# also assigns weight to each college based on how far the tuition is from the range the user selected
# same format as sat_match()
def tuition_match(bin_val):
    tuition_dict_weights = {}
    #print(dict_tuit_schools_out)
    for key, value in dict_tuit_schools_in.items():
        if bin_val == "VOID":
            tuition_dict_weights[key] = 15
        else:
            # if school is in same state as user, use in-state tuition for weight
            if user_state_live == dict_state_schools[key]:
                set_bin = dict_tuit_schools_in[key]
                
                if bin_val == set_bin:
                    tuition_dict_weights[key] = 100
                elif ((bin_val + 1) == set_bin) or (bin_val - 1 == set_bin):
                    tuition_dict_weights[key] = 75
                elif ((bin_val + 2) == set_bin) or (bin_val - 2 == set_bin):
                    tuition_dict_weights[key] = 60
                elif ((bin_val + 3) == set_bin) or (bin_val - 3 == set_bin):
                    tuition_dict_weights[key] = 35
                elif ((bin_val + 4) == set_bin) or (bin_val - 4 == set_bin):
                    tuition_dict_weights[key] = 20
                elif ((bin_val + 5) == set_bin) or (bin_val - 5 == set_bin):
                    tuition_dict_weights[key] = 5
                elif ((bin_val + 6) == set_bin) or (bin_val - 6 == set_bin):
                    tuition_dict_weights[key] = 0
                else:
                    tuition_dict_weights[key] = 0
            # otherwise, use out-of-state tuition for weight
            else:
                set_bin = dict_tuit_schools_out[key]
                
                if bin_val == set_bin:
                    tuition_dict_weights[key] = 100
                elif ((bin_val + 1) == set_bin) or (bin_val - 1 == set_bin):
                    tuition_dict_weights[key] = 75
                elif ((bin_val + 2) == set_bin) or (bin_val - 2 == set_bin):
                    tuition_dict_weights[key] = 60
                elif ((bin_val + 3) == set_bin) or (bin_val - 3 == set_bin):
                    tuition_dict_weights[key] = 35
                elif ((bin_val + 4) == set_bin) or (bin_val - 4 == set_bin):
                    tuition_dict_weights[key] = 20
                elif ((bin_val + 5) == set_bin) or (bin_val - 5 == set_bin):
                    tuition_dict_weights[key] = 5
                elif ((bin_val + 6) == set_bin) or (bin_val - 6 == set_bin):
                    tuition_dict_weights[key] = 0
                else:
                    tuition_dict_weights[key] = 0
                
    return tuition_dict_weights
```

## MATCHMAKER    
    

```python
import collections, functools, operator
from operator import itemgetter

# assign a final cumulative weight for each college based on user input
def matchmaker(sat, act, gpa, acr, academic_weight, enroll, campsize, setting, state, enroll_weight, tuit, pnm, cost_weight, 
               division, athletic_weight, majors_list, cost_major):
    sat_val = {}
    act_val = {}
    gpa_val = {}
    acr_val = {}
    enroll_val = {}
    campsize_val = {}
    setting_val = {}
    state_val = {}
    tuit_val = {}
    pnm_val = {}
    major_val = {}
    division_val = {}
    
    # assign a weight to each statistic for each college based on the pre-determined weight of that statistic in the category
    # and the weight of the category inputted by the user
    # 100 means it is a match with the user's preference
    for key, value in sat.items():
        # sat_val[key] = value * WGT1_SAT * WGT_ACADEM
        sat_val[key] = value * WGT1_SAT * academic_weight * 0.01
                
    for key, value in act.items():
        # act_val[key] = value * WGT1_ACT * WGT_ACADEM
        act_val[key] = value * WGT1_ACT * academic_weight * 0.01
                
    for key, value in gpa.items():
        # gpa_val[key] = value * WGT1_GPA * WGT_ACADEM
        gpa_val[key] = value * WGT1_GPA * academic_weight * 0.01
    
    for key, value in acr.items():
        # acr_val[key] = value * WGT1_ACR * WGT_ACADEM
        acr_val[key] = value * WGT1_ACR * academic_weight * 0.01 
    
    for key, value in enroll.items():
        # enroll_val[key] = value * WGT2_ENRO * WGT_ENROLL
        enroll_val[key] = value * WGT2_ENRO * enroll_weight * 0.01
                
    for key, value in campsize.items():
        # campsize_val[key] = value * WGT2_CAMP * WGT_ENROLL
        campsize_val[key] = value * WGT2_CAMP * enroll_weight * 0.01
                
    for key, value in setting.items():
        setting_val[key] = value * WGT2_SETT * enroll_weight * 0.01
        
    for key, value in state.items():
        state_val[key] = value * WGT2_STATE * enroll_weight * 0.01
        
    for key, value in tuit.items():
        # tuit_val[key] = value * WGT3_TUIT * WGT_COST
        tuit_val[key] = value * WGT3_TUIT * cost_weight * 0.01
                
    for key, value in pnm.items():
        # pnm_val[key] = value * WGT3_PNM * WGT_COST
        pnm_val[key] = value * WGT3_PNM * cost_weight * 0.01
        
    for key, value in majors_list.items():
        major_val[key] = value * WGT4_MAJ * cost_major * 0.01
        
    for key, value in division.items():
        # division_val[key] = value * WGT5_DIV * WGT_ATH
        division_val[key] = value * WGT5_DIV * athletic_weight * 0.01
        
    
    #------------------------------------------------------------------------------------------------
        
    # add weight of each statistic together for each college to get one final cumulative weight for each college
    list1= [sat_val, act_val, gpa_val, acr_val, enroll_val, campsize_val, setting_val, state_val, major_val, tuit_val, pnm_val, division_val]
    result = dict(functools.reduce(operator.add, map(collections.Counter, list1)))
    
    # formatting result dict
    min_diff = 100
    for key, value in result.items():
        diff = (100 - value) * (4/5)
        if diff < min_diff:
            min_diff = diff
        
    for key, value in result.items():
        if value >= 100:
            result[key] = 99.7
        else: 
            result[key] = round(min_diff + value, 2)
    
    return result
```
